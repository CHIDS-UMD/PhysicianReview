```python
import numpy as np
import pandas as pd
import json


def change_dict_names(d, replace_name_list):
    d2 = {}
    for k, v in d.items():
        for item in replace_name_list:
            k = k.replace(item, '')
        d2[k] = v
    return d2

def change_cred(x):
    
    if type(x) == str:
        try:
            x = x.replace(' ', '').replace('.', '').replace(';', '')
            x = x[:-1] if x[-1] == ',' else x
            return x
        except:
            return None
    else:
        return None
    
    
def check_empty_dict(d):
    return bool(sum([pd.isnull(v) for k, v in d.items()]))


def get_clean_data_info(data):
    if '_' not in [i for i in data][0]:
        L = {k:v for k,v in data.items() if not pd.isnull(v)}
        
    else:
        L = []
        for i in list(set([int(i.split('_')[-1]) for i in data])):
            # print(i)
            d = {k:data[k] for k in data if '_{}'.format(i) in k}
            d = {k.split('_')[0]:data[k] for k in data if int(k.split('_')[-1]) == i}
            if check_empty_dict(d) == False:
                L.append(d) 
            else:
                break

    return None if len(L) == 0 else L #json.dumps(L, indent = 2)


def get_clean_row_data(row, newName2Cols, replace_name_list):
    new_row = {}
    for newname, cols in newName2Cols.items():
        # print(newname)
        # print(row[cols])

        if 'Info' in newname or 'Address' in newname:
            # print('\n{}'.format(newname))
            data = row[cols].to_dict()

            clean_data = get_clean_data_info(data)
            # print(clean_data)
            # print(type(clean_data))
            new_row[newname] = clean_data
        
        elif 'Credential' in newname:
            new_row[newname] = change_cred(row[cols])
            
        else:
            # print
            new_row[newname] = row[cols] if not pd.isnull(row[cols]) else None

    # L.append(new_row)
    
    
    new_row2 = {}
    for k, v in new_row.items():
        if v == None:
            continue
        elif 'Date' in k:
            new_row2[k] = pd.to_datetime(v)
        elif 'Info' in k or 'Address' in k:
            if type(v) == dict:
                v = change_dict_names(v, replace_name_list)
            elif type(v) == list:
                v = [change_dict_names(d, replace_name_list) for d in v]
            new_row2[k] = json.dumps(v)
        elif 'EntityType' in k:
            new_row2[k] = int(v)
        else:
            new_row2[k] = v
            
    # main TaxonomyInfo
    if 'TaxonomyInfo' in new_row2:
        TaxonomyInfo = json.loads(new_row2['TaxonomyInfo'])
        if len(TaxonomyInfo) > 0:
            MainTaxonomy = TaxonomyInfo[0]["Taxonomy Code"]
            new_row2['MainTaxonomy'] = MainTaxonomy
        
    return new_row2



new_row = get_clean_row_data(row, newName2Cols, replace_name_list)
new_row# ['EntityType'] != 1
```

# SQL Schema

```python
import os
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, String, Integer, Table, Text, DateTime
from sqlalchemy import ForeignKey
from sqlalchemy.orm import relationship
from sqlalchemy.orm import sessionmaker

def sdb_connect(basedir, name = 'NEEPS'):
    return create_engine('sqlite:///'+os.path.join(basedir, name + '.sqlite'))

def mdb_connect():
    MySQL_DB = 'mysql+pymysql://root:@localhost:3306/physicianinfo?charset=utf8'
    return create_engine(MySQL_DB)


Base = declarative_base()

def create_table(engine):
    Base.metadata.create_all(engine)

    
class NPPES(Base):
    __tablename__ = 'NPPES'
    id = Column(Integer, primary_key=True)
    NPI = Column(Integer)
    EntityType = Column(Integer)
    ReplacementNPI = Column(Text)
    EIN = Column(Text)
    
    Organization = Column(Text)
    LastName = Column(Text)
    FirstName= Column(Text)
    MiddleName  = Column(Text)
    Prefix = Column(Text)
    Suffix = Column(Text)
    Gender = Column(Text)
    Credential= Column(Text)
    Sole = Column(Text)
    MainTaxonomy = Column(Text)
    
    CertificationDate  = Column(DateTime)
    EnumerationDate = Column(DateTime)
    LastUpdateDate = Column(DateTime)
    
    BusinessMailingAddress = Column(Text)
    PracticeLocationAddress= Column(Text)
    
    OfficialInfo= Column(Text)
    OtherNameInfo  = Column(Text)
    AffiliationInfo = Column(Text)
    
    DeactivationInfo  = Column(Text)
    TaxonomyInfo = Column(Text)
    TaxonomyGroupInfo= Column(Text)
    OtherIdentifierInfo  = Column(Text)
    
```
